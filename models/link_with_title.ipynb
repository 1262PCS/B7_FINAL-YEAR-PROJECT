```python
import pandas as pd
import re
```

```python
df = pd.read_csv('/content/cora/papers', sep='\t', header=None, names=['paper_id', 'word_attributes', 'class_label'])
```

```python
pattern = r"<title>(.*?)</title>"

# Extract the title using apply with a lambda function
def extract_title(text):
  if not isinstance(text, str):
      return ""  # Return empty string for non-strings
  match = re.search(pattern, text)
  return match.group(1) if match else None
df['title'] = df['class_label'].apply(extract_title)
df
```

```python
import dgl
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from dgl.nn import GraphConv
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
import torch.nn.functional as F


# Function to load the Cora dataset from DGL
def load_cora_data():
    data = dgl.data.CoraGraphDataset()
    graph = data[0]
    features = graph.ndata['feat']
    labels = graph.ndata['label']
    train_mask = graph.ndata['train_mask']
    test_mask = graph.ndata['test_mask']
    return graph, features, labels, train_mask, test_mask

# Function to preprocess the data
def preprocess_data(graph, features, labels, train_mask, test_mask):
    graph = dgl.add_self_loop(graph)
    num_feats = features.shape[1]
    num_classes = len(torch.unique(labels))
    return graph, features, labels, train_mask, test_mask, num_feats, num_classes

# Function to create a simple Graph Convolutional Network (GCN) model for link prediction
class LinkPredictionModel(nn.Module):
    def __init__(self, in_feats, hidden_size, out_feats):
        super(LinkPredictionModel, self).__init__()
        self.conv1 = GraphConv(in_feats, hidden_size)
        self.conv2 = GraphConv(hidden_size, out_feats)

    def forward(self, graph, features):
        x = F.relu(self.conv1(graph, features))
        x = F.relu(self.conv2(graph, x))
        return x

# Function to train the link prediction model
def train_model(model, graph, features, train_mask, labels, criterion, optimizer, num_epochs=100):
    for epoch in range(num_epochs):
        model.train()
        logits = model(graph, features)
        loss = criterion(logits[train_mask], labels[train_mask])
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {loss.item()}')

# Function to evaluate the link prediction model
def evaluate_model(model, graph, features, test_mask, labels):
    model.eval()
    with torch.no_grad():
        logits = model(graph, features)
    predicted_labels = torch.argmax(logits[test_mask], dim=1)
    true_labels = labels[test_mask]
    accuracy = accuracy_score(true_labels.numpy(), predicted_labels.numpy())
    print(f'accuracy :',accuracy)
```

```python
# Main function
def main():
    graph, features, labels, train_mask, test_mask = load_cora_data()
    graph, features, labels, train_mask, test_mask, num_feats, num_classes = preprocess_data(
        graph, features, labels, train_mask, test_mask)

    # Define the link prediction model
    model = LinkPredictionModel(num_feats, 64, num_classes)

    # Define the loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Train the link prediction model
    train_model(model, graph, features, train_mask, labels, criterion, optimizer, num_epochs=50)

    # Evaluate the link prediction model
    evaluate_model(model, graph, features, test_mask, labels)

    # Take input node ID from the user
    input_node_id = int(input("Enter the input node ID: "))

    # Get the neighbors (cited nodes) of the input node
    neighbors = list(graph.predecessors(input_node_id).numpy())
    cited_node_titles = df.loc[neighbors, 'title'].tolist()
    print(f'Nodes cited by node {input_node_id}: {neighbors}')
    print(f'Titles of cited nodes: {cited_node_titles}')

if __name__ == "__main__":
    main()
```
